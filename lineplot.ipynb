```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1, 2, 3, 4, 5, 6, 7]
var1 = [2, 3, 4, 1, 6, 7, 8]
x = pd.DataFrame({"var": var, "var1": var1})

sns.lineplot(x='var', y='var1', data=x)

plt.show()
```

```python
plt.plot(var,var1)
plt.show()
```

```python
y= sns.load_dataset("penguins")
y
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y)

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y,hue="sex")

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y,hue="sex",size=10)

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y,hue="sex",size=20,style="sex")

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y,hue="sex",size=20,style="sex",palette="Reds")

plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y,hue="sex",size=20,style="sex",palette="Blues",markers=["o",">"])

plt.show()
```
